```python
from PIL import Image
import io
import os
import numpy as np
import cv2
import time


def prompted_segmentation(image_path, mode="point", point=None, label=None, box=None, save_dir="outputs"):
    """
    Performs SAM segmentation with manual prompt using Ultralytics SAM model.

    Args:
        image_path: Path to the image file.
        mode: 'auto' | 'point' | 'box'
        point: list of [x, y] points (e.g., [[x, y]]) for positive/negative prompts
        label: list of int labels (1 for foreground, 0 for background) matching `point`
        box: tuple/list (x1, y1, x2, y2) for a single bounding box prompt
        save_dir: directory to save output mask

    Returns:
        (orig_image_pil, binary_mask_np_uint8, overlay_image_pil)
    """
    image_path = os.path.abspath(image_path)
    os.makedirs(save_dir, exist_ok=True)

    if not os.path.exists(image_path):
        raise FileNotFoundError(f"Image not found: {image_path}")

    if mode not in {"auto", "point", "box"}:
        raise ValueError("Mode must be 'auto', 'point', or 'box'")

    if mode == "point":
        if point is None or label is None:
            raise ValueError("For 'point' mode, both 'point' and 'label' must be provided")
        if len(point) != len(label):
            raise ValueError("'point' and 'label' lengths must match")

    if mode == "box":
        if box is None or len(box) != 4:
            raise ValueError("For 'box' mode, 'box' must be a (x1, y1, x2, y2) tuple/list")

    image_pil = Image.open(image_path).convert("RGB")

    t0 = time.time()
    if mode == "auto":
        results = model.predict(source=image_path)
    elif mode == "point":
        results = model.predict(source=image_path, points=point, labels=label)
    elif mode == "box":
        results = model.predict(source=image_path, bboxes=[box])
    infer_time = time.time() - t0

    r = results[0]
    combined = None
    try:
        masks = r.masks.data
        masks_np = masks.cpu().numpy() if hasattr(masks, "cpu") else np.asarray(masks)
        combined = (np.any(masks_np > 0, axis=0).astype(np.uint8) * 255)
    except Exception:
        h, w = r.orig_shape[:2] if hasattr(r, "orig_shape") else image_pil.size[::-1]
        combined = np.zeros((h, w), dtype=np.uint8)
        if getattr(r, "masks", None) and getattr(r.masks, "xy", None):
            for poly in r.masks.xy:
                pts = np.array(poly, dtype=np.int32)
                if pts.size > 0:
                    cv2.fillPoly(combined, [pts], 255)

    base = os.path.splitext(os.path.basename(image_path))[0]
    mask_path = os.path.join(save_dir, base + "_mask.png")
    Image.fromarray(combined).save(mask_path)
    print(f" Inference Time: {infer_time:.2f}s | Mask saved at: {mask_path}")

    overlay = np.array(image_pil).copy()
    overlay[combined == 255] = [255, 0, 0]
    overlay_img = Image.fromarray(overlay)

    return image_pil, combined, overlay_img
```

```python
def clean_mask(mask_uint8, min_area=500):
    """
    Keeps only the largest connected white area (main object)
    """
    mask_bin = (mask_uint8 > 127).astype(np.uint8)
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(mask_bin, connectivity=8)

    if num_labels <= 1:
        return mask_uint8

    areas = stats[1:, cv2.CC_STAT_AREA]
    largest_idx = np.argmax(areas) + 1
    cleaned = (labels == largest_idx).astype(np.uint8) * 255
    return cleaned

mask_clean = clean_mask(mask)
Image.fromarray(mask_clean).save("outputs/cleaned_mask.png")
plt.imshow(mask_clean, cmap='gray')
plt.title("Cleaned Mask (Foreground Only)")
plt.axis('off')
plt.show()
```

```python
def apply_mask(image_path, mask_path, save_path="outputs/segmented_object.png"):
    img = Image.open(image_path).convert("RGBA")
    mask = Image.open(mask_path).convert("L")

    img_arr = np.array(img)
    mask_arr = np.array(mask)

    img_arr[..., 3] = mask_arr
    result = Image.fromarray(img_arr)
    result.save(save_path)
    print(f" Saved isolated object at: {save_path}")
    return result

isolated = apply_mask("E:\Coding\PROGRAMS\Deep learning Projects\Image-Segmentation\download (3).jpeg", "outputs/cleaned_mask.png")
plt.imshow(isolated)
plt.gca().set_facecolor('black')
plt.axis('off')
plt.title("Isolated Object (Transparent BG)")
plt.show()
```

```python
def show_results(title, results):
    # Support both tuple output from prompted_segmentation and dict-based outputs
    if isinstance(results, tuple):
        img = results[2]
    elif isinstance(results, dict):
        overlay_val = results.get("overlay")
        img = overlay_val if isinstance(overlay_val, Image.Image) else Image.open(overlay_val)
    else:
        raise TypeError("Unsupported results type for show_results")
    plt.figure(figsize=(6,4))
    plt.imshow(img)
    plt.axis('off')
    plt.title(title)
    plt.show()
```
